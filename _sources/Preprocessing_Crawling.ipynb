# Preprocessing Data Text

Langkah awal dalam analisis teks dan pemrosesan bahasa alami (NLP) adalah tahapan preprocessing data teks. Preprocessing bertujuan membersihkan, mengubah, dan mengorganisir data teks agar lebih sesuai untuk analisis atau penggunaan dalam model pembelajaran mesin. Berikut beberapa tahap umum preprocessing data teks yang sering digunakan di Google Colab:

1. **Pembersihan Teks (Text Cleaning):**
   - Menghapus Karakter Khusus: Menghilangkan tanda baca, simbol, atau karakter tidak relevan.

2. **Tokenisasi:**
  - Memecah teks menjadi token-token atau kata-kata kecil. Tokenisasi dapat melibatkan pemisahan berdasarkan spasi atau tanda baca.

3. **Menghilangkan Kata Stop (Stopword Removal):**
   - Menghapus kata-kata umum yang tidak memberikan nilai informasi tinggi.

4. **Stemming atau Lemmatization:**
  - Mengubah kata-kata menjadi bentuk dasarnya. Stemming memotong akhiran kata, sementara lemmatization mengembalikan kata ke bentuk dasarnya.

5. **Vektorisasi:**
   - Mengonversi kata atau token menjadi representasi numerik menggunakan metode seperti TF-IDF atau Word Embeddings (Word2Vec, GloVe).

6. **Pengelompokan (Text Categorization):**
   - Mengkategorikan teks ke dalam kelas atau kategori, umumnya digunakan dalam tugas klasifikasi teks.

7. **Pemisahan Data:**
   - Membagi dataset menjadi set pelatihan, validasi, dan pengujian.

8. **Pengujian dan Evaluasi:**
   - Menguji dan mengevaluasi hasil preprocessing untuk memastikan data teks siap digunakan dalam analisis atau pemodelan.

Preprocessing data teks sangat penting dalam NLP dan dapat memengaruhi performa model dan analisis. Oleh karena itu, memahami dan menerapkan dengan cermat berbagai tahap preprocessing sangat dianjurkan.

## Cleaning Data

### Menghapus Karakter Tertentu

Fungsi `cleaning` berfungsi membersihkan teks yang terdapat dalam kolom 'Abstrak'. Proses pembersihan ini mencakup penghapusan karakter-karakter yang tidak relevan seperti tanda baca, serta mengonversi teks menjadi daftar kata-kata.

````python
def cleaning(text):
  text = re.sub(r'[^a-zA-Z\s]', '', text).strip()
  return text

df['Cleaning'] = df['Abstrak'].apply(cleaning)
df['Cleaning']
````

Fungsi `cek_specialCharacter` difungsikan untuk mengidentifikasi keberadaan karakter khusus dalam teks yang telah melalui proses pembersihan. Apabila karakter khusus terdeteksi, teks tersebut akan dicetak.

````python
def cek_specialCharacter(dokumen):
  karakter = ['!', '@', '#', '$', '%', '^', '&', '*', '(', ')', '-', '_', '+', '=', '{', '}', '[', ']', '|', '\\', ':', ';', '"', "'", '<', '>', ',', '.', '?', '/', '`', '~']
  for i in dokumen:
    if i in karakter :
      print(dokumen)
df['Cleaning'].apply(cek_specialCharacter)
````

## Tokenizing

Fungsi `tokenizer` berperan dalam melakukan tokenisasi dan proses case folding pada teks yang telah melalui tahap pembersihan. Proses ini mengubah teks menjadi token-token kata.

````python
def tokenizer(text):
  text = text.lower()
  return word_tokenize(text)

df['Tokenizing'] = df['Cleaning'].apply(tokenizer)
df['Tokenizing']
````

Menghitung jumlah kata dalam tiap abstrak

````python
def count_word(dokumens):
  return len(dokumens)

df['Count Word'] = df['Tokenizing'].apply(count_word)
df
````

## Stopword

Stopword adalah kata-kata umum yang biasanya tidak memberikan nilai signifikan dalam analisis teks. Fungsi `stopwordText` dipakai untuk mengeliminasi stopword dari token-token kata yang sudah diperoleh. Setelah token-token kata dibersihkan dan stopword dihapus, mereka kemudian digabungkan kembali menjadi teks utuh dan disimpan dalam kolom 'Full Text'.

````python
corpus = stopwords.words('indonesian')

def stopwordText(words):
 return [word for word in words if word not in corpus]

df['Stopword Removal'] = df['Tokenizing'].apply(stopwordText)

# Gabungkan kembali token menjadi kalimat utuh
df['Full Text'] = df['Stopword Removal'].apply(lambda x: ' '.join(x))
df['Full Text']
````

## One Hot Encoding

### Fungsi One Hot Encoder

Fungsi `pandasOneHotEncoder` dimanfaatkan untuk menerapkan one-hot encoding pada token-token kata yang sudah melewati proses pembersihan dan eliminasi stopword. Outputnya berupa DataFrame yang mencerminkan keberadaan atau ketiadaan setiap kata dalam setiap dokumen.

````python
def pandasOneHotEncoder(dokumens, label):
  encoder  = pd.get_dummies(dokumens.apply(pd.Series).stack()).sum(level=0)
  df = pd.concat([dokumens, encoder, label], axis=1)

  return df

oneHotEncoder = pandasOneHotEncoder(df['Stopword Removal'], df['Label'])
oneHotEncoder
````

````python
oneHotEncoder.to_csv('OneHotEncoder.csv', index=False)
````

## TF IDF

### Fungsi TF IDF

Fungsi `tfidf` berfungsi untuk menerapkan vektorisasi TF-IDF pada teks yang telah melalui proses pembersihan dan eliminasi stopword. Hasilnya adalah representasi numerik dari dokumen menggunakan skema TF-IDF.

````python
def tfidf(dokumen, label):
  vectorizer = TfidfVectorizer()
  x = vectorizer.fit_transform(dokumen).toarray()
  terms = vectorizer.get_feature_names_out()

  final_tfidf = pd.DataFrame(x, columns=terms)
  final_tfidf.insert(0, 'Dokumen', dokumen)
  final_tfidf.insert(len(final_tfidf.columns),'Label', label)

  return (vectorizer, final_tfidf)

tfidf_vectorizer, final_tfidf = tfidf(df['Full Text'], df['Label'])
final_tfidf
````

### Save ke CSV

````python
final_tfidf.to_csv('TF-IDF.csv', index=False)
````

## Term Frequensi

### Fungsi Term Frequensi

Fungsi `term_freq` berperan dalam menjalankan vektorisasi Term Frequency (TF) pada teks yang telah melalui proses pembersihan dan eliminasi stopword. Outputnya berupa representasi numerik dari dokumen dengan memanfaatkan skema Term Frequency.

````python
def term_freq(dokumens, label):
  # Buat objek CountVectorizer
  vectorizer = CountVectorizer()
  tf_matrix = vectorizer.fit_transform(dokumens).toarray()
  terms = vectorizer.get_feature_names_out()

  final_tf = pd.DataFrame(tf_matrix, columns=terms)
  final_tf.insert(0, 'Dokumen', dokumens)
  final_tf.insert(len(final_tf.columns),'Label', label)

  return (vectorizer, final_tf, tf_matrix, terms)

tf_vectorizer, final_tf, tf_matrix, tf_terms = term_freq(df['Full Text'], df['Label'])
final_tf
````


## Logarithm Freqency

### Fungsi Logarithm Frequensi

Fungsi `logarithm_freq` berfungsi untuk melakukan transformasi frekuensi logaritmik pada data Term Frequency. Tujuannya adalah untuk mengurangi dampak dominasi kata-kata yang sangat umum dalam analisis teks.

````python
def logarithm_freq(dokumens):
  return np.log10(dokumens + 1)

df_logarithm_freq = pd.DataFrame(tf_matrix, columns=tf_terms).apply(logarithm_freq)
df_logarithm_freq.insert(0, 'Dokumen', df['Full Text'])
df_logarithm_freq.insert(len(df_logarithm_freq.columns),'Label', df['Label'])
df_logarithm_freq
````
